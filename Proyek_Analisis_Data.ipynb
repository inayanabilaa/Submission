## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Wrangling

### Gathering Data

```python
customer_data = pd.read_csv('/content/customers_dataset.csv')
payment_data = pd.read_csv('/content/order_payments_dataset.csv')
print(payment_data)
```

**Insight:**
- Dataset yang digunakan dalam proyek ini 2 dataset customer dataset dan order payments dataset
- Ketika melakukan print dataset, user dapat melihat jumlah rows dan column yang dimiliki oleh dataset

### Assessing Data

```python
print("Informasi Customer:")
print(customer_data.info())

print("Informasi Pembayaran:")
print(payment_data.info())

print("Informasi Customer:")
print(customer_data.describe())

print("Informasi Pembayaran:")
print(payment_data.describe())
```

**Insight:**
- pembayaran paling tinggi bernilai 13664.08.
- Kota dan negara bagian pelanggan dapat digunakan untuk analisis geografis atau demografi.

### Cleaning Data

```python
print("Memeriksa missing value customer data:")
print(customer_data.isnull().sum())

print("Memeriksa duplikat customer data:")
print(customer_data.duplicated().sum())

print("Memeriksa missing value payment data:")
print(payment_data.isnull().sum())

print("Memeriksa duplikat payment data:")
print(payment_data.duplicated().sum())
```

**Insight:**
- Tidak terdapat missing value pada data
- Tidak terdapat duplicate data

## Exploratory Data Analysis (EDA)

```python
top_cities = customer_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Kota dengan Jumlah Pelanggan Terbanyak', fontsize=16)
plt.xlabel('Jumlah Pelanggan', fontsize=12)
plt.ylabel('Kota', fontsize=12)
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(data=payment_data, x='payment_installments', kde=False, bins=20, color='purple')
plt.title('Distribusi Jumlah Cicilan', fontsize=16)
plt.xlabel('Jumlah Cicilan', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.show()
```

**Insight:**
- Pelanggan paling banyak berada pada kota sao paulo
- Berdasarkan data pembayaran, customer jarang menggunakan cicilan untuk melakukan pembayaran atau pembayaran dengan 0 cicilan

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
top_cities = customer_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Kota dengan Jumlah Pelanggan Terbanyak', fontsize=16)
plt.xlabel('Jumlah Pelanggan', fontsize=12)
plt.ylabel('Kota', fontsize=12)
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=payment_data, x='payment_type', order=payment_data['payment_type'].value_counts().index, palette='pastel')
plt.title('Distribusi Metode Pembayaran', fontsize=16)
plt.xlabel('Metode Pembayaran', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.show()
```

**Insight:**
- Customer terbanyak berasal dari kota Sao Paulo
- Metode pembayaran yang paling banyak digunakan adalah Credit Card
